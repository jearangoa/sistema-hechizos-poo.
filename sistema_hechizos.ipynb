```python
# Definimos clase Hechizo y en dificultad, las opciones son baja, media, alta
class Hechizo:
    def __init__(self, nombre, efecto, dificultad):
        self.nombre = nombre
        self.efecto = efecto
        self.dificultad = dificultad  

    def __str__(self):
        return f"Hechizo: {self.nombre} (Efecto: {self.efecto}, Dificultad: {self.dificultad})"

# Definimos clase Estudiante y en casa, las opciones son Gryffindor, Slytherin, Ravenclaw, Hufflepuff
class Estudiante:
    def __init__(self, nombre, casa):
        self.nombre = nombre
        self.casa = casa  
        self.hechizos_aprendidos = []

    def aprender_hechizo(self, hechizo):
        if hechizo not in self.hechizos_aprendidos:
            self.hechizos_aprendidos.append(hechizo)
            return f"{self.nombre} ha aprendido el hechizo '{hechizo.nombre}'."
        else:
            return f"{self.nombre} ya conoce el hechizo '{hechizo.nombre}'."

    def listar_hechizos(self):
        if self.hechizos_aprendidos:
            return f"Hechizos aprendidos por {self.nombre}:\n" + "\n".join(str(h) for h in self.hechizos_aprendidos)
        else:
            return f"{self.nombre} no ha aprendido ningún hechizo aún."

    def __str__(self):
        return f"Estudiante: {self.nombre} (Casa: {self.casa})"

# Definimos clase EscuelaDeMagia
class EscuelaDeMagia:
    def __init__(self, nombre):
        self.nombre = nombre
        self.estudiantes = []
        self.hechizos = []

    def agregar_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)
        return f"Estudiante {estudiante.nombre} ha sido registrado en {self.nombre}."

    def agregar_hechizo(self, hechizo):
        self.hechizos.append(hechizo)
        return f"Hechizo '{hechizo.nombre}' ha sido agregado a la lista de {self.nombre}."

    def listar_estudiantes(self):
        return "\n".join(str(e) for e in self.estudiantes)

    def listar_hechizos(self):
        return "\n".join(str(h) for h in self.hechizos)

# este seria el ejemplo
if __name__ == "__main__":
    # hechizos
    expelliarmus = Hechizo("Expelliarmus", "Desarma al oponente", "Media")
    wingardium_leviosa = Hechizo("Wingardium Leviosa", "Levita objetos", "Baja")
    avada_kedavra = Hechizo("Avada Kedavra", "Mata instantáneamente", "Alta")

    # estudiantes
    harry = Estudiante("Harry Potter", "Gryffindor")
    hermione = Estudiante("Hermione Granger", "Gryffindor")
    draco = Estudiante("Draco Malfoy", "Slytherin")

    #escuela de magia que en este ejemplo sera
    hogwarts = EscuelaDeMagia("Hogwarts")

    # Agregamos estudiantes y hechizos a la escuela
    print(hogwarts.agregar_estudiante(harry))
    print(hogwarts.agregar_estudiante(hermione))
    print(hogwarts.agregar_estudiante(draco))

    print(hogwarts.agregar_hechizo(expelliarmus))
    print(hogwarts.agregar_hechizo(wingardium_leviosa))
    print(hogwarts.agregar_hechizo(avada_kedavra))

    # Estudiantes aprenden hechizos
    print(harry.aprender_hechizo(expelliarmus))
    print(hermione.aprender_hechizo(wingardium_leviosa))
    print(draco.aprender_hechizo(avada_kedavra))

    # Listamos estudiantes y hechizos
    print("\nEstudiantes en Hogwarts:")
    print(hogwarts.listar_estudiantes())

    print("\nHechizos disponibles en Hogwarts:")
    print(hogwarts.listar_hechizos())

    # Mostramos los hechizos aprendidos
    print("\n" + harry.listar_hechizos())
    print("\n" + hermione.listar_hechizos())
    print("\n" + draco.listar_hechizos())
```
